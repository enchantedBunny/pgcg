```python
import pyCalc as p
import matplotlib.pyplot as plt
one = p.var_old('c', 1)# 1
two = p.var_old('c', 2)# 1
six = p.var_old('c', 6)# 1
twtyfour = p.var_old('c', 24)# 1
mone = p.var_old('c', -1) # -1
x = p.var_old('i');
unm = p.var_old('f');
unm.f(mone,x,'*') # -x
unex = p.var_old('f')
unex.exp(unm) #e**-x
fn1 = p.var_old('f')
fn1.f(one, unex, '+')#1+e**-x
fn2 = p.var_old('f')
fn2.f(one, fn1, '/') #1/ (1+ e**-x)
xx = [i/10 for i in range(-40,50)]
ys = [fn2.value([i, 0]) for i in xx]
plt.plot(xx,ys)
plt.title('Sigmoid')
```

```python
plt.show()
```

```python
xxx = [i/10 for i in range(-40,50)]
yxs = [fn2.getDeriv(0,[i]) for i in xx]
plt.plot(xxx,yxs)
plt.title('Sigmoid deriv')
```

```python
plt.show()
```

```python
y = p.var_old()
y.pow(x, 8)
xx = [i/10 for i in range(-50,51)]
ys = [y.value([i, 0]) for i in xx]
plt.plot(xx,ys)
plt.title('x**8')
```

```python
plt.show()
```

```python
ys = [y.getDeriv(0,[i]) for i in xx]
plt.plot(xx,ys)
plt.title('derivative of x**8')
```
